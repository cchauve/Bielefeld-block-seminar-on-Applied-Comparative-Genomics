We can define groups of species:
- The Gambia complex: AQUAS, AARAD, AMELC, AMERM, AGAMP
- two outgroups to the Gambia complex (not forming a clade): ACHRA, AEPIE
- three groups of asian/european mosquitoes (plue one african): AMINM, ACULA, AFUNF, AMACM, ASTES and AFARF, ADIRW and ASINS, AATRE,
- a group of American mosquitoes: ADARC, AALBS

```python
# Creating groups of species
SP_GAMBIA_IG = ['AQUAS', 'AARAD', 'AMELC', 'AMERM', 'AGAMP']
SP_GAMBIA_OG = ['ACHRA', 'AEPIE'] # Warning: These two species do not form a clade
SP_AFRICAN   = SP_GAMBIA_IG+SP_GAMBIA_OG
SP_ASIAN_1   = ['AMINM', 'ACULA', 'AFUNF', 'AMACM', 'ASTES']
SP_ASIAN_2   = ['AFARF', 'ADIRW']
SP_ASIAN_3   = ['ASINS', 'AATRE']
SP_CELLIA    = SP_AFRICAN+SP_ASIAN_1+SP_ASIAN_2
SP_AMERICAN  = ['ADARC', 'AALBS']
SP_ALL       = SP_AMERICAN+SP_ASIAN_3+SP_CELLIA
SP_NB        = len(SP_ALL)
```

## Genome assemblies

We can now look at the genome assemblies, especially to the fragmentation level and the distribution of the length of the contigs. We look at two cases, considering all contigs and considering only contigs with at least one gene.

```python
# Creating a contigs dataframe
# The information associated to each contig is its species, its length and the number of genes it contains
# Here we assume that no two contigs have the same name.

# Contigs length files suffix
ctg_lg_suffix = '_ctg_lg.txt'
# All genes files suffix
genes_all_suffix = '_genes_all.txt'

# Recording species, length and genes present in each contig
CTG_DATA  = {}
for species in SP_ALL:
    ctg_lg_data = open('../data/'+species+ctg_lg_suffix,'r').readlines()
    for ctg_lg in ctg_lg_data:
        if ctg_lg[0]!='#':
            ctg = ctg_lg.split()[0]
            lg  = int(ctg_lg.split()[1])
            CTG_DATA[ctg] = {}
            CTG_DATA[ctg]['lg']=lg
            CTG_DATA[ctg]['species']=species
            CTG_DATA[ctg]['genes']=[]
    genes_data = open('../data/'+species+genes_all_suffix,'r').readlines()
    for gene in genes_data:
        if gene[0]!='#':
            ctg  = gene.split()[1]
            name = gene.split()[2]
            if ctg!='MT': # We do not have the mitochondrial contigs
                CTG_DATA[ctg]['genes'].append(name)
                
DF_CTG_DATA = pd.DataFrame.from_dict(CTG_DATA,orient='index',columns=['species','lg','genes'])
```

```python
# Genome length by contigs, binned.

# Bins for contigs length
CTG_LG_BINS_VAL    = [1000,10000,100000,1000000,10000000]
CTG_LG_BINS_LABELS = ['<=1kb','<=10kb','<=100kb','<=1Mb','<=10Mb','>10Mb']
CTG_LG_BINS_NB     = len(CTG_LG_BINS_VAL)

#Recording the cumulated length of contigs in each bin per species
CTG_LG_W_GENES = {species: [0 for ctg_lg in CTG_LG_BINS_LABELS] for species in SP_ALL} # contigs with a gene
CTG_LG_ALL     = {species: [0 for ctg_lg in CTG_LG_BINS_LABELS] for species in SP_ALL} # call contigs

for species in SP_ALL:
    ctg_lg_data  = DF_CTG_DATA[DF_CTG_DATA['species']==species]['lg']
    ctg_genes_data = DF_CTG_DATA[DF_CTG_DATA['species']==species]['genes']
    for ctg in ctg_lg_data.keys():
        lg  = ctg_lg_data[ctg]
        for lg_ind in range(CTG_LG_BINS_NB):
            if lg<=CTG_LG_BINS_VAL[lg_ind] and (lg_ind==0 or lg>CTG_LG_BINS_VAL[lg_ind-1]):
                CTG_LG_ALL[species][lg_ind]+=lg
                if len(ctg_genes_data[ctg])>0:
                    CTG_LG_W_GENES[species][lg_ind]+=lg
        if lg>CTG_LG_BINS_VAL[CTG_LG_BINS_NB-1]:
            CTG_LG_ALL[species][CTG_LG_BINS_NB]+=lg
            if len(ctg_genes_data[ctg])>0:
                CTG_LG_W_GENES[species][CTG_LG_BINS_NB]+=lg
```

```python
# Figure

# Stacked bat chart parameters
CTG_LG_INDEX = np.arange(SP_NB)
barWidth = 0.35

plt.figure(figsize=(15, 7))
# Figure 1: all contigs
plt.subplot(121)
# Transpose of CTG_LG_BINS
CTG_LG_ALL_BINS_TRANSP = {B:[CTG_LG_ALL[sp][B] for sp in SP_ALL] for B in range(CTG_LG_BINS_NB+1)}
plt.bar(CTG_LG_INDEX, CTG_LG_ALL_BINS_TRANSP[0], width=barWidth)
barBottom = np.add([0 for species in SP_ALL],CTG_LG_ALL_BINS_TRANSP[0])
for B in range(1,CTG_LG_BINS_NB+1):
    plt.bar(CTG_LG_INDEX, CTG_LG_ALL_BINS_TRANSP[B], bottom=barBottom, width=barWidth)
    barBottom = np.add(barBottom,CTG_LG_ALL_BINS_TRANSP[B])
plt.ylabel('Cumulated length')
plt.title('Genome size by binned contigs size\nall contigs')
plt.xticks(CTG_LG_INDEX,SP_ALL,rotation=45)
#plt.yticks(np.arange(0, 81, 10))
plt.legend(CTG_LG_BINS_LABELS)

# Figure 2: contigs with genes
plt.subplot(122)
# Transpose of CTG_LG_BINS
CTG_LG_W_GENES_BINS_TRANSP = {B:[CTG_LG_W_GENES[sp][B] for sp in SP_ALL] for B in range(CTG_LG_BINS_NB+1)}
plt.bar(CTG_LG_INDEX, CTG_LG_W_GENES_BINS_TRANSP[0], width=barWidth)
barBottom = np.add([0 for species in SP_ALL],CTG_LG_W_GENES_BINS_TRANSP[0])
for B in range(1,CTG_LG_BINS_NB+1):
    plt.bar(CTG_LG_INDEX, CTG_LG_W_GENES_BINS_TRANSP[B], bottom=barBottom, width=barWidth)
    barBottom = np.add(barBottom,CTG_LG_W_GENES_BINS_TRANSP[B])
plt.ylabel('Cumulated length')
plt.title('Genome size by binned contigs size\n contigs with at least one gene')
plt.xticks(CTG_LG_INDEX,SP_ALL,rotation=45)

plt.show()
```

#### Comments.

So it is clear that some of the genomes are very poorly assembled. This is particularly the case for AMELC, ACHRA, ACULA and AMACM, and to a lesser extent, for ADARC and ASINS, where we see close to no genes in contigs over 1Mb. 

This suggests it might be interesting to consider a reduced dataset without these species.

## Genes

Next, we look at the genes. We start by looking at the cumulated length of both the full genes and the exons. We also look at the number of genes and exons, as well as the number of contigs containing genes.

```python
# Genes data frame.
# Information associated to each gene: species, contig, sign, coordinates, exons coordinates

GENES_DATA = {}
for species in SP_ALL:
    genes_data = open('../data/'+species+genes_all_suffix,'r').readlines()
    for gene in genes_data:
        if gene[0]!='#':
            exons1= gene.rstrip().split()[7].split(':')
            exons = []
            for exon in exons1:
                exon2 = exon.split('-')
                exons.append((int(exon2[0]),int(exon2[1])))
            species = gene.split()[0]
            ctg   = gene.split()[1]
            name  = gene.split()[2]
            sign  = gene.split()[3]
            start = int(gene.split()[4])
            end   = int(gene.split()[5])
            
            GENES_DATA[name] = {'species':species,'ctg':ctg,'sign':sign,'start':start,'end':end,'exons':exons}
                      
DF_GENES_DATA = pd.DataFrame.from_dict(GENES_DATA,orient='index')
```

```python
SUMMARY_GENES = {species: {'nb_ctgs':0,'lg_full':0,'lg_exons':0,'lg_ratio':0,'nb_full':0,'nb_exons':0,'avg_lg_full':0,'avg_lg_exons':0} for species in SP_ALL}
for species in SP_ALL:
    #genes_data = open('../data/'+species+genes_all_suffix,'r').readlines()
    genes_data  = DF_GENES_DATA[DF_GENES_DATA['species']==species].transpose()
    ctgs = set()
    for gene in genes_data.keys():
        exons = genes_data[gene]['exons']
        ctg   = genes_data[gene]['ctg']
        g_start = genes_data[gene]['start']
        g_end   = genes_data[gene]['end']
        ctgs.add(ctg)
        for exon in exons:
            start = exon[0]
            end   = exon[1]
            SUMMARY_GENES[species]['lg_exons']+=end-start+1
            SUMMARY_GENES[species]['nb_exons']+=1
        SUMMARY_GENES[species]['lg_full']+=g_end-g_start+1
        SUMMARY_GENES[species]['nb_full']+=1
    SUMMARY_GENES[species]['nb_ctgs']=len(ctgs)
    SUMMARY_GENES[species]['lg_ratio']=SUMMARY_GENES[species]['lg_full']/SUMMARY_GENES[species]['lg_exons']
    SUMMARY_GENES[species]['avg_lg_full']=SUMMARY_GENES[species]['lg_full']/SUMMARY_GENES[species]['nb_full']
    SUMMARY_GENES[species]['avg_lg_exons']=SUMMARY_GENES[species]['lg_exons']/SUMMARY_GENES[species]['nb_exons']
```

```python
COL_SUMMARY_GENES_DF=['nb_ctgs','lg_full','nb_full','avg_lg_full','lg_exons','nb_exons','avg_lg_exons','lg_ratio']

print('Explanation of the column labels:')
print('\tnb_ctgs: number of contigs containing at least one gene')
print('\tlg_full: cumulated length of the genes, including introns')
print('\tnb_full: number of genes')
print('\tavg_full: average length of a gene')
print('\tlg_exons: cumulated length of the exons')
print('\tnb_exons: number of exons')
print('\tavg_lg_exons: average length of an exon')
print('\tlg_ratio: lg_full/lg_exons\n')
pd.DataFrame.from_dict(SUMMARY_GENES,orient='index',columns=COL_SUMMARY_GENES_DF)
```

#### Comments

We can see some significant variations. 
- AGAMP, the only fully assembled genome has a larger length of genes, but we see other less well assembled species with similar cumulated genes length, AFARF, AMERM and ASINS, the last two being quite less well assembled.
- There is no super-clear correlation between average gene length and cumulated gene length.
- AMACM has both the smallest average gene length and the largest number of genes; could-it be that many annotated genes are actually only exons.
- Conversely, ACHRA has the second smallest number of genes and the second smallest average gene length, suggesting that the long genes have not been assembled, which is consistant with the assembly statistics.
- But ADARC has also the smallest number of genes, but a normal average gene length, although a smaller number of exons.

In fine, there is no simple explanation of the relation between assembly fragmentation and gene statistics.

## Gene families

We look now at the gene families. The data are provided as one RaXML gene tree per family and they were obtained from <a href="http://people.csail.mit.edu/waterhouse/ODBMOZ2/trees_ALL.tar.gz">ODBMOZ2/trees_ALL.tar.gz</a>.

```python
# Gene families dictionary: fam -> list of genes

# File containing a gene tree per family with at least two genes
raxml_data = open('../data/RAxML_bestTrees.txt','r').readlines()
FAM2GENES_ALL  = {}
FAM2GENES_SUB1 = {}
FAM_ALL_EMPTY  = []
FAM_SUB1_EMPTY = []

for fam_data in raxml_data:
    if fam_data[0]=='M':
        family = fam_data.rstrip()
    else:
        tree = fam_data.replace('(',' ').replace(')',' ').replace(',',' ').replace(':',' ')
        FAM2GENES_ALL[family]  = []
        FAM2GENES_SUB1[family] = []
        for w in tree.split():
            if w[0]=='A' and 'ASTEI' not in w: 
            # Likely data error due to including an extra species of An. stephensi, as we have no ASTEI gene in our data
                FAM2GENES_ALL[family].append(w)
                if DF_GENES_DATA.loc[w]['species'] in SP_SUB1:
                    FAM2GENES_SUB1[family].append(w)
        if len(FAM2GENES_ALL[family])==0:
            FAM2GENES_ALL.pop(family,None)
            FAM_ALL_EMPTY.append(family)
        if len(FAM2GENES_SUB1[family])==0:
            FAM2GENES_SUB1.pop(family,None)
            FAM_SUB1_EMPTY.append(family)
FAM_ALL_NAMES  = list(FAM2GENES_ALL.keys())
FAM_SUB1_NAMES = list(FAM2GENES_SUB1.keys())
```

```python
print('\n')
print('Number of non-empty families, all species: '+str(len(FAM_ALL_NAMES)))
print('Number of empty families, all species: '+str(len(FAM_ALL_EMPTY)))
print('Number of non-empty families, selected species: '+str(len(FAM_SUB1_NAMES)))
print('Number of empty families, selected species: '+str(len(FAM_SUB1_EMPTY)))
```

```python
# Gene families dictionary: fam -> list of species

FAM2SPECIES_ALL  = {} # For all species
FAM2SPECIES_SUB1 = {} # For the selected species

for family in FAM_ALL_NAMES:
    FAM2SPECIES_ALL[family]  = set() # Using set is slow
    FAM2SPECIES_SUB1[family] = set()
    genes = FAM2GENES_ALL[family]
    for gene in genes:
        species = DF_GENES_DATA.loc[gene]['species']
        FAM2SPECIES_ALL[family].add(species)
        if species in SP_SUB1:
            FAM2SPECIES_SUB1[family].add(species)
    FAM2SPECIES_ALL[family]  = list(FAM2SPECIES_ALL[family])
    FAM2SPECIES_SUB1[family] = list(FAM2SPECIES_SUB1[family])
    if len(FAM2SPECIES_SUB1[family])==0:
        FAM2SPECIES_SUB1.pop(family,None)
```

```python
# Distribution of the number of species covered by each family, all species considered
FAM_ALL_SPECIES_SIZE = [len(FAM2SPECIES_ALL[fam]) for fam in FAM2SPECIES_ALL.keys()]
FAM_ALL_SPECIES_MAX  = max(FAM_ALL_SPECIES_SIZE)
# Distribution of the number of genes per species
FAM_ALL_GENES_SIZE = [len(FAM2GENES_ALL[fam]) for fam in FAM2GENES_ALL.keys()]
FAM_ALL_GENES_MAX  = max(FAM_ALL_GENES_SIZE)
```

```python
# Distribution of the number of species covered by each family, selected species
FAM_SUB1_SPECIES_SIZE = [len(FAM2SPECIES_SUB1[fam]) for fam in FAM2SPECIES_SUB1.keys()]
FAM_SUB1_SPECIES_MAX  = max(FAM_SUB1_SPECIES_SIZE)
# Distribution of the number of genes per species
FAM_SUB1_GENES_SIZE = [len(FAM2GENES_SUB1[fam]) for fam in FAM2GENES_SUB1.keys()]
FAM_SUB1_GENES_MAX  = max(FAM_SUB1_GENES_SIZE)
```

```python
# Figure

plt.figure(figsize=(15, 7))

plt.subplot(121)
plt.hist(FAM_ALL_SPECIES_SIZE, bins=np.arange(FAM_ALL_SPECIES_MAX+2), align='left', rwidth=0.5)
plt.title('Number of covered species of gene families, all species')
plt.ylabel('Number of families')
plt.xlabel('Number of covered species')
plt.xticks(np.arange(FAM_ALL_SPECIES_MAX+1))

plt.subplot(122)
plt.hist(FAM_SUB1_SPECIES_SIZE, bins=np.arange(FAM_SUB1_SPECIES_MAX+2), align='left', rwidth=0.5)
plt.title('Number of covered species of gene families, selected species')
plt.ylabel('Number of families')
plt.xlabel('Number of covered species')
plt.xticks(np.arange(FAM_SUB1_SPECIES_MAX+1))
plt.show()
```

We can observe a much higher number of families covering all species in the selected subset dataset, compared to the dataset containing all species. This is not surprising, but again suggests the selected species dataset is cleaner.

We can also observe, in both cases, the some kind of symmetry that suggests that several families have been split into at least two families.

## Genes overlap

One of the issues we face for gene order studies on genomes whose genes can be composed of several exons is that genes sometimes overlap. We explore this issue with the *Anopheles* data.

```python
DF_GENES_DATA_SORTED_BY_POS = DF_GENES_DATA.sort_values(['species','ctg','start'],ascending=[1,1,1]).transpose()

OVERLAPS_SAME_STRAND  = {species: [] for species in SP_ALL}
OVERLAPS_DIFF_STRANDS = {species: [] for species in SP_ALL}

(gene_prev,species_prev,ctg_prev,sign_prev,end_prev)=('','','','',0)
for gene in DF_GENES_DATA_SORTED_BY_POS:
    species = DF_GENES_DATA_SORTED_BY_POS[gene]['species']
    ctg     = DF_GENES_DATA_SORTED_BY_POS[gene]['ctg']
    sign    = DF_GENES_DATA_SORTED_BY_POS[gene]['sign']
    start   = DF_GENES_DATA_SORTED_BY_POS[gene]['start']
    end     = DF_GENES_DATA_SORTED_BY_POS[gene]['end']
    if species==species_prev and ctg==ctg_prev and start<end_prev:
        if sign==sign_prev:
            OVERLAPS_SAME_STRAND[species].append((gene_prev,gene))
        else:
            OVERLAPS_DIFF_STRANDS[species].append((gene_prev,gene))
    (gene_prev,species_prev,ctg_prev,sign_prev,end_prev)=(gene,species,ctg,sign,end)
# Note: we overlook potential overlaps between more than 2 genes
```

```python
NB_OVERLAPS_SAME_STRAND  = {species: len(OVERLAPS_SAME_STRAND[species]) for species in SP_ALL}
NB_OVERLAPS_DIFF_STRANDS = {species: len(OVERLAPS_DIFF_STRANDS[species]) for species in SP_ALL}
print('Number of gene overlaps on the same strand: ')
print(NB_OVERLAPS_SAME_STRAND)
print('Number of gene overlaps on different strands: ')
print(NB_OVERLAPS_DIFF_STRANDS)
```

#### Comments.
We can observe a large number of overlaps between pairs of genes located on different strands, corrsponding to more than the gene complement of a single species, with some species having more then 10% of their genes involved in overlaps.
